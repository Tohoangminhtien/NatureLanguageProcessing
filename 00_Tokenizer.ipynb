# Tokenizer

Tokenizer là một phương pháp tách từ cơ bản trong NLP, hoạt động bằng cách tạo 1 từ điển bao hàm tất cả corpus/sequence với key là text, values là một số nguyên riêng biệt. Khi mô hình hóa 1 đoạn văn bản với tokenizer thì ta lấy từ điển đã tạo ánh xạ với đoạn văn bản thì sẽ được một dãy số tương ứng, với những từ không có trong từ điển sẽ được gán chung vào 1 key gọi là Over Of Vocabulary OOV. <br>
Để các đầu ra có độ dài bằng nhau có thể dùng pad sequence tương tự padding trong CNN

```python
from sklearn.feature_extraction.text import CountVectorizer
from keras.preprocessing.text import Tokenizer

# Dữ liệu văn bản
corpus=df['title'].tolist()
```

```python
token=Tokenizer(oov_token='<OOV>')
token.fit_on_texts(corpus)
```

```python
word_frequency=token.word_index
word_count=token.word_counts
```

```python
word_count_sorted = sorted(word_count.items(), key=lambda x: x[1],reverse=True)
word_count_sorted
```

## Text to sequence with corpus

```python
corpus=token.texts_to_sequences(corpus)
```

```python
corpus_padded=pad_sequences(corpus)
corpus_padded.shape
```

## Training sentiment classifier with Tokenizer

```python
X=corpus_padded
y=df['subject']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score

y_pred=model.predict(X_test)
acc=accuracy_score(y_test,y_pred)
# f1=f1_score(y_test,y_pred)
```

```python
print(f'Accuracy: {acc}')
```

```python
# Input from keyboard
text_pred = input('Nhap tieu de cua news: ')
while (len(text_pred.split(' ')) > 21):
    text_pred = input('Nhap tieu de cua news: ')

# Processing
text_pred = [text_pred]
text_pred_sequence = token.texts_to_sequences(text_pred)
text_pred_padded = pad_sequences(text_pred_sequence, maxlen=21)
print(text_pred_padded)

# Predict
result_predict = model.predict(text_pred_padded)
result_predict
```
